# 1. Objective <a id="objective"></a>

[&#10148; Back to the table of contents](#table_of_contents)

Along the steps 1 and 2 of this tutorial, we have collected all the necessary data (step 1), and checked for structural inconsistencies (step 2).
We will now bring some adjustments and reduce model's complexity.

$\bullet$ <strong>Section 3:</strong><br/>First, we will rename metabolites to fit GBA formalism (external metabolites must have a prefix `x_`; old suffices are removed).

$\bullet$ <strong>Section 4:</strong><br/>We will replace the default `RNA` molecule by ribosomal RNA (`rRNA`) including a realistic synthesis reaction, as it is assumed that most RNA in the minimal cell is ribosomal ([Breuer et al. 2019](https://elifesciences.org/articles/36842)).

$\bullet$ <strong>Section 5:</strong><br/>Then, we will operate an important simplification of the amino-acid and tRNA metabolism by reducing these metabolites to single pseudo-molecules, which will facilitate further the solving and the analysis of the model.

$\bullet$ <strong>Sections 6 and 7:</strong><br/>Finally, a few major modifications are required to obtain a working GBA model:
- Create or edit the ribosomal reaction for protein synthesis. This reaction must be named `Ribosome` and produce the metabolite `Protein`. A protein synthesis reaction already exists in the MMSYN model: `Protein_transl`. However, we will have to edit the enzyme composition to fit the ribosomal composition.
- A few reactions should be removed, because they will never be used. This is the case of protein degradation (`Protein_degrad`) and the transport of alternative sugars (the standard minimal cell medium is always based on glucose).

# 3. <a id="rename"></a> Rename all metabolites based on GBA formalism

[&#10148; Back to the table of contents](#table_of_contents)

<div align="center">

| GBApy method | Description |
| ------ | ----------- |
| <code>gba.GbaBuilder.rename_metabolite( previous_id, new_id )</code> | Rename a metabolite |

</div>

To fit GBA formalism, we must rename external and internal metabolites:
- External metabolites must have a prefix `x_`; we will also remove the suffix `_e`,
- We will remove the suffix `_c` for internal metabolites.

```python
previous_ids = list(builder.metabolites.keys())
for m_id in previous_ids:
    new_id = m_id
    if new_id.endswith("_e"):
        new_id = "x_"+new_id[:-2]
        builder.rename_metabolite(m_id, new_id)
    elif new_id.endswith("_c"):
        new_id = new_id[:-2]
        builder.rename_metabolite(m_id, new_id)
```

The list of rRNAs is available in the genome assembly. We will calculate an average rRNA.

We must then edit the biosynthesis reaction. Following [Breuer et al. (2019)](https://elifesciences.org/articles/36842), we assume a cost of <strong>0.41 ATP/nucleotide</strong> in RNA synthesis.

```python
# Build the average ribosomal RNA
RNA_file = SeqIO.parse(open("data/ASM170832v2/GCA_001708325.2_ASM170832v2_rna_from_genomic.fna", "r"), "fasta")
n_atp    = 0.0
n_ctp    = 0.0
n_gtp    = 0.0
n_utp    = 0.0
count    = 0.0
for record in RNA_file:
    if "ribosomal RNA" in record.description:
        seq     = str(record.seq).replace("T", "U")
        n_atp  += seq.count("A")
        n_ctp  += seq.count("C")
        n_gtp  += seq.count("G")
        n_utp  += seq.count("U")
        count  += 1.0
n_atp = int(n_atp/count)
n_ctp = int(n_ctp/count)
n_gtp = int(n_gtp/count)
n_utp = int(n_utp/count)
average_seq = "A" * n_atp + "C" * n_ctp + "G" * n_gtp + "U" * n_utp
rRNA = gba.Metabolite(id="rRNA", name="Average ribosomal RNA",
                      species_location=gba.SpeciesLocation.Internal, species_type=gba.SpeciesType.RNA,
                      formula=average_seq)
rRNA.calculate_mass()
builder.add_metabolite(rRNA)
builder.metabolites["rRNA"].summary()

# Build the ribosomal RNA polymerase reaction
n_ppi    = n_atp+n_ctp+n_gtp+n_utp
atp_cost = int(n_ppi*0.41)
stoichiometry = {
    "atp": -n_atp-atp_cost,
    "ctp": -n_ctp,
    "gtp": -n_gtp,
    "utp": -n_utp,
    "h2o": -atp_cost,
    "rRNA": 1,
    "adp": atp_cost,
    "h": atp_cost,
    "pi": atp_cost,
    "ppi": n_ppi
}
reaction = gba.Reaction(id="rRNAP", name="rRNA polymerase reaction",
                        reaction_type=gba.ReactionType.Metabolic,
                        lb=0.0, ub=1000.0,
                        metabolites=stoichiometry,
                        proteins=builder.reactions["RNAP"].proteins,
                        GPR=builder.reactions["RNAP"].GPR,
                        enzyme_mass=builder.reactions["RNAP"].enzyme_mass)
reaction.enforce_kcat_irreversibility()
reaction.enforce_km_irreversibility()
reaction.complete(kcat_value=45000.0, km_value=0.00013)
builder.add_reaction(reaction)
builder.reactions["rRNAP"].summary()

# Build the ribosomal RNA degradation reaction
stoichiometry = {
    "rRNA": -1,
    "h2o": -atp_cost*2.3621, # Manual adjustment to keep mass balance
    "amp": n_atp,
    "cmp": n_ctp,
    "gmp": n_gtp,
    "ump": n_utp,
    "h": atp_cost,
}
reaction = gba.Reaction(id="rRNA_degrad", name="rRNA degradation reaction",
                        reaction_type=gba.ReactionType.Metabolic,
                        lb=0.0, ub=1000.0,
                        metabolites=stoichiometry,
                        proteins=builder.reactions["RNA_degrad"].proteins,
                        GPR=builder.reactions["RNA_degrad"].GPR,
                        enzyme_mass=builder.reactions["RNA_degrad"].enzyme_mass)
reaction.enforce_kcat_irreversibility()
reaction.enforce_km_irreversibility()
reaction.complete(kcat_value=45000.0, km_value=0.00013)
builder.add_reaction(reaction)
builder.reactions["rRNA_degrad"].summary()
```

## 5.3. Calculate average amino-acids and tRNAs <a id="AA_edit_3"></a>

[&#10148; Back to the table of contents](#table_of_contents)

<div align="center">

| GBApy method | Description |

We now have to calculate:
- An average amino-acid,
- An average uncharged tRNA from tRNA sequences,
- An average charged tRNA by summing the mass of the average uncharged tRNA and the average amino-acid.

```python
# Calculate an average amino-acid
average_AA_mass = 0.0
for aa_id in internal_amino_acids:
    average_AA_mass += builder.metabolites[aa_id].mass
average_AA_mass /= len(internal_amino_acids)

x_AA = gba.Metabolite(id="x_AA", name="Average external amino-acid",
                      species_location=gba.SpeciesLocation.External,
                      species_type=gba.SpeciesType.SmallMolecule,
                      formula="", mass=average_AA_mass)
AA = gba.Metabolite(id="AA", name="Average internal amino-acid",
                    species_location=gba.SpeciesLocation.Internal,
                    species_type=gba.SpeciesType.SmallMolecule,
                    formula="", mass=average_AA_mass)
builder.add_metabolite(x_AA)
builder.add_metabolite(AA)
builder.metabolites["x_AA"].summary()
builder.metabolites["AA"].summary()

# Calculate average uncharged/charged tRNAs
n_atp = 0.0
n_ctp = 0.0
n_gtp = 0.0
n_utp = 0.0
count = 0.0
for m_id in uncharged_trnas:
    seq     = builder.metabolites[m_id].formula
    n_atp  += seq.count("A")
    n_ctp  += seq.count("C")
    n_gtp  += seq.count("G")
    n_utp  += seq.count("U")
    count  += 1.0
n_atp       = int(n_atp/count)
n_ctp       = int(n_ctp/count)
n_gtp       = int(n_gtp/count)
n_utp       = int(n_utp/count)
average_seq = "A" * n_atp + "C" * n_ctp + "G" * n_gtp + "U" * n_utp
tRNA = gba.Metabolite(id="tRNA", name="Average uncharged tRNA",
                      species_location=gba.SpeciesLocation.Internal,
                      species_type=gba.SpeciesType.RNA,
                      formula=average_seq)
tRNA.calculate_mass()
AAtRNA = gba.Metabolite(id="AAtRNA", name="Average charged tRNA",
                        species_location=gba.SpeciesLocation.Internal,
                        species_type=gba.SpeciesType.RNA,
                        formula="", mass=tRNA.mass+average_AA_mass)
builder.add_metabolite(tRNA)
builder.add_metabolite(AAtRNA)
builder.metabolites["tRNA"].summary()
builder.metabolites["AAtRNA"].summary()
```

```python
# Average amino-acid transporter
stoichiometry = {
    "x_AA": -1,
    "atp": -1,
    "h2o": -1,
    "AA": 1,
    "adp": 1,
    "h": 1,
    "pi": 1
}
reaction = gba.Reaction(id="AAabc", name="Average amino-acid transporter",
                        reaction_type=gba.ReactionType.Transport,
                        lb=0.0, ub=1000.0,
                        metabolites=stoichiometry,
                        proteins=builder.reactions["ALAt2r"].proteins,
                        GPR=builder.reactions["ALAt2r"].GPR,
                        enzyme_mass=builder.reactions["ALAt2r"].enzyme_mass)
reaction.enforce_kcat_irreversibility()
reaction.enforce_km_irreversibility()
reaction.complete(kcat_value=45000.0, km_value=0.00013)
builder.add_reaction(reaction)
builder.reactions["AAabc"].summary()

# Average tRNA synthesis
seq      = builder.metabolites["tRNA"].formula
n_atp    = seq.count("A")
n_ctp    = seq.count("C")
n_gtp    = seq.count("G")
n_utp    = seq.count("U")
n_ppi    = n_atp+n_ctp+n_gtp+n_utp
atp_cost = int(n_ppi*0.41)
stoichiometry = {
    "atp": -n_atp-atp_cost,
    "ctp": -n_ctp,
    "gtp": -n_gtp,
    "utp": -n_utp,
    "h2o": -atp_cost,
    "tRNA": 1,
    "adp": atp_cost,
    "h": atp_cost,
    "pi": atp_cost,
    "ppi": n_ppi
}
reaction = gba.Reaction(id="tRNAP", name="tRNA polymerase reaction",
                        reaction_type=gba.ReactionType.Metabolic,
                        lb=0.0, ub=1000.0,
                        metabolites=stoichiometry,
                        proteins=builder.reactions["rRNAP"].proteins,
                        GPR=builder.reactions["rRNAP"].GPR,
                        enzyme_mass=builder.reactions["rRNAP"].enzyme_mass)
reaction.enforce_kcat_irreversibility()
reaction.enforce_km_irreversibility()
reaction.complete(kcat_value=45000.0, km_value=0.00013)
builder.add_reaction(reaction)
builder.reactions["tRNAP"].summary()

# Average tRNA charging reaction
stoichiometry = {
    "AA": -1,
    "tRNA": -1,
    "atp": -1,
    "AAtRNA": 1,
    "amp": 1,
    "ppi": 1
}
reaction = gba.Reaction(id="AATRS", name="Average tRNA charging reaction",
                        reaction_type=gba.ReactionType.Metabolic,
                        lb=0.0, ub=1000.0,
                        metabolites=stoichiometry,
                        proteins=builder.reactions["ALATRS"].proteins,
                        GPR=builder.reactions["ALATRS"].GPR,
                        enzyme_mass=builder.reactions["ALATRS"].enzyme_mass)
reaction.enforce_kcat_irreversibility()
reaction.enforce_km_irreversibility()
reaction.complete(kcat_value=45000.0, km_value=0.00013)
builder.add_reaction(reaction)
builder.reactions["AATRS"].summary()
```

## 5.6. Remove or edit dependent pathways <a id="AA_edit_6"></a>

[&#10148; Back to the table of contents](#table_of_contents)

| GBApy method | Description |
| ------ | ----------- |

We have to reconstruct the amino-acid metabolism by:
- Deleting pathways becoming useless or unapplicable after the simplification,
- Editing existent reactions including amino-acids or tRNAs.

Here are the pathways to delete:

# 6. Edit proteins synthesis reaction <a id="protein_edit"></a>

[&#10148; Back to the table of contents](#table_of_contents)

- <strong>First</strong>, we rename the reaction `Protein_transl` into `Ribosome`, and its product `protein` into `Protein`, to match GBA formalism.
- <strong>Second</strong>, we edit the enzyme composition to make the production of the ribosomal machinery more realistic. To do so, we will collect all the ribosomal proteins and tell the model that all proteins are required to build the enzyme (AND GPR association).
- <strong>Third</strong>, we will delete the protein degradation, which does not match GBA formalism.

```python

# Edit names to match the GBA formalism
builder.rename_reaction("Protein_transl", "Ribosome")
builder.rename_metabolite("protein", "Protein")

# Edit the ribosome reaction enzyme composition
ribosome     = builder.reactions["Ribosome"]
ribosome.GPR = gba.ReactionGPR.AND
ribosome.proteins.clear()
for p in builder.proteins.values():
    if not p.product is None and "ribosomal protein" in p.product:
        ribosome.add_proteins({p.id: 1})
ribosome.calculate_enzyme_mass()
ribosome.summary()

# Remove protein degradation
builder.remove_reaction("Protein_degrad")
```
